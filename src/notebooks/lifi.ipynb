# Get all the chains

```python
import requests
import json
import asyncio
import pandas as pd

async def get_chains():
    # chainTypes can be of type SVM and EVM, by default, only EVM chains will be returned
    optional_chain_type = "EVM"
    url = 'https://li.quest/v1/chains'

    params = {'chainTypes': optional_chain_type}

    try:
        response = await asyncio.to_thread(requests.get, url, params=params)
        response.raise_for_status()  # Raise an exception for HTTP errors (e.g., 404, 500)
        data = response.json()
        return data
    except requests.exceptions.RequestException as e:
        # Handle request-related errors (e.g., network issues, timeout)
        print(f"Error: {e}")
        return None

# Fetch the chains data using asyncio
chains_data = await get_chains()

if chains_data:
    # Save the chains data as JSON
    with open('chains_data.json', 'w') as json_file:
        json.dump(chains_data, json_file, indent=4)
    print("Chains data saved as 'chains_data.json'")
else:
    print("Failed to fetch chains data.")
```

```python
if chains_data:

    chains_df = pd.json_normalize(chains_data['chains'])
    chains_df.to_csv('chains_data.csv', index=False)

chains_df
```

# Get routes for a specific pathway

```python
async def get_routes(fromAmount, fromChainId, fromTokenAddress, fromAddress, toChainId, toTokenAddress):

    url = "https://li.quest/v1/advanced/routes"

    params = {
        "allowDestinationCall": True,
        "fromAmount": fromAmount,
        "fromChainId": fromChainId,
        "fromTokenAddress": fromTokenAddress,
        "fromAddress": fromAddress,
        "toChainId": toChainId,
        "toTokenAddress": toTokenAddress
    }
    headers = {
        "accept": "application/json",
        "content-type": "application/json"
    }

    try:
        response = requests.post(url, json=params, headers=headers)
        response.raise_for_status()  # Raise an exception for HTTP errors (e.g., 404, 500)
        data = response.json()
        return data
    except requests.exceptions.RequestException as e:
        # Handle request-related errors (e.g., network issues, timeout)
        print(f"Error: {e}")
        return None
    
def save_routes_data(routes_data,from_chain,to_chain,token_name,amount,datetime):
    import os

    folder_path = 'routes'

    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    if routes_data:
        with open(f'{folder_path}/{from_chain}-{to_chain}-{token_name}.json', 'w') as json_file:
            json.dump(routes_data, json_file, indent=4)
        print(f"Routes data saved as {folder_path}/{from_chain}-{to_chain}-{token_name}-{amount}-{datetime}.json", 'w')
    else:
        print(f"Failed to fetch routes data.{from_chain}-{to_chain}-{token_name}-{amount}", "w")

def df_routes_data(routes_data, from_chain,to_chain,token_name,amount, datetime):
    import pandas as pd
    import os

    folder_path = 'routes'

    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    # Initialize an empty DataFrame for the normalized data
    normalized_data_df = pd.DataFrame()
    raw_data_df = pd.DataFrame()

    # Loop through each route in the data
    for route in routes_data['routes']:
        # Extract the route metadata (excluding 'steps')
        metadata = {k: v for k, v in route.items() if k != 'steps'}
        metadata_df = pd.json_normalize(metadata)
        metadata_df.columns = ['route_' + col for col in metadata_df.columns]  # Add a prefix to each metadata column name

        # Loop through each step in the route
        for step in route['steps']:
            # Normalize the step data (excluding 'feeCosts' and 'gasCosts')
            step_data = {k: v for k, v in step.items() if k not in ['feeCosts', 'gasCosts']}
            step_df = pd.json_normalize(step_data)
            step_df.columns = ['step_' + col for col in step_df.columns]  # Add a prefix to each metadata column name

            # Combine step data with route metadata
            combined_df = pd.concat([step_df] * len(metadata_df), ignore_index=True)
            combined_df = pd.concat([combined_df, metadata_df], axis=1)

            # Normalize and append fee costs data if present
            if step['estimate']['feeCosts']:
                for fee in step['estimate']['feeCosts']:
                    fee_df = pd.json_normalize(fee)
                    fee_df['cost_type'] = 'feeCost'  # Add a cost type column to distinguish fee costs
                    fee_combined_df = pd.concat([combined_df.iloc[[0]], fee_df], axis=1)
                    normalized_data_df = pd.concat([normalized_data_df, fee_combined_df], ignore_index=True)

            # Normalize and append gas costs data if present
            if step['estimate']['gasCosts']:
                for gas in step['estimate']['gasCosts']:
                    gas_df = pd.json_normalize(gas)
                    gas_df['cost_type'] = 'gasCost'  # Add a cost type column to distinguish gas costs
                    gas_combined_df = pd.concat([combined_df.iloc[[0]], gas_df], axis=1)
                    normalized_data_df = pd.concat([normalized_data_df, gas_combined_df], ignore_index=True)

            # If there are no feeCosts or gasCosts, just append the step data
            if not(step['estimate']['feeCosts']) and not(step['estimate']['gasCosts']):
                normalized_data_df = pd.concat([normalized_data_df, combined_df], ignore_index=True)
    

    # Save the DataFrame to a CSV file
    normalized_data_df.to_csv(f'{folder_path}/{from_chain}-{to_chain}-{token_name}-{amount}-{datetime}.csv', index=False, escapechar='\\')
    
    return normalized_data_df

def generate_pathways(pathwaychains_df, tokens):
    from itertools import product

    pathways = []
    pathways_df = pathwaychains_df
    pathways_df = pd.merge(pathways_df,chains_df[['id','key','name']],left_on='fromChainId',right_on='id')
    pathways_df.rename(columns={'name': 'fromChainName', 'key': 'fromChainKey'}, inplace=True)
    pathways_df = pd.merge(pathways_df,chains_df[['id','key','name']],left_on='toChainId',right_on='id')
    pathways_df.rename(columns={'name': 'toChainName', 'key': 'toChainKey'}, inplace=True)
    pathways_df.drop(columns=['id_y','id_x'], inplace=True)

    # Create a Cartesian product of the DataFrame and the list of tokens
    cartesian_product = pd.DataFrame(list(product(pathways_df['toChainName'], pathways_df['toChainKey'], tokens)), columns=['toChainName', 'toChainKey', 'tokenName'])

    # Concatenate the Cartesian product DataFrame with the existing pathways_df DataFrame
    pathways_df = pd.concat([pathways_df] * len(tokens), ignore_index=True)

    # Add the 'tokenName' column to the existing DataFrame with values from the Cartesian product
    pathways_df['tokenName'] = cartesian_product['tokenName']

    pathways_df = pd.merge(pathways_df,tokens_df[['address','chainId','symbol','decimals','name','coinKey','priceUSD']],left_on=['fromChainId','tokenName'],right_on=['chainId','symbol'])
    pathways_df.drop(columns=['chainId','symbol'], inplace=True)
    pathways_df.rename(columns={'address': 'fromTokenAddress', 'decimals': 'fromDecimals','name': 'fromName','coinKey': 'fromCoinKey'}, inplace=True)
    pathways_df = pd.merge(pathways_df,tokens_df[['address','chainId','symbol','decimals','name','coinKey']],left_on=['toChainId','tokenName'],right_on=['chainId','symbol'])
    pathways_df.drop(columns=['chainId','symbol'], inplace=True)
    pathways_df.rename(columns={'address': 'toTokenAddress', 'decimals': 'toDecimals','name': 'toName','coinKey': 'toCoinKey'}, inplace=True)


    pathways_df.to_csv('pathways.csv', index=False, escapechar='\\')

    pathways = pathways_df.to_dict(orient='records')

    '''for row in chains_df.itertuples(index=False):
        print(row.Name, row.Age)

    for x in y:
        pathway = {}
        pathway['fromAmount']
        pathway['fromChainId']
        pathway['fromTokenAddress']
        pathway['fromAddress']
        pathway['toChainId']
        pathway['toTokenAddress']
        pathway['fromChainName']
        pathway['toChainName']
        pathway['tokenName']'''

    return pathways

async def get_multiple_routes(pathways):
    from datetime import datetime
    import os

    folder_path = 'routes'

    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    pathways_df = pd.DataFrame()

    for pathway in pathways:
        for i in [1,2,3,4,5,6]:
            fromChainId = pathway['fromChainId']
            fromAmount = int(float((10 ** (pathway['fromDecimals'])) * (10 ** i)) /  float(pathway['priceUSD']))
            fromTokenAddress = pathway['fromTokenAddress']
            fromAddress = pathway['fromTokenAddress']
            toChainId = pathway['toChainId']
            toTokenAddress = pathway['toTokenAddress']
            fromChainName = pathway['fromChainName']
            toChainName = pathway['toChainName']
            tokenName = pathway['tokenName']

            currentdatetime = datetime.now()

            routes_data = await get_routes(fromAmount, fromChainId, fromTokenAddress, fromAddress, toChainId, toTokenAddress)
            save_routes_data(routes_data,fromChainName,toChainName,tokenName,fromAmount,currentdatetime)
            if routes_data:
                pathway_df = df_routes_data(routes_data,fromChainName,toChainName,tokenName,fromAmount,currentdatetime)
                pathways_df = pd.concat([pathways_df,pathway_df], ignore_index=True)
    
    pathways_df.to_csv(f'{folder_path}/pathways-{currentdatetime}.csv', index=False, escapechar='\\')

    return pathways_df
```

```python
import pandas as pd
import json

# Load the JSON data from the file
file_path = 'routes/ETH-ARB-USDT.json'
with open(file_path, 'r') as file:
    data = json.load(file)

# Initialize an empty DataFrame for the normalized data
normalized_data_df = pd.DataFrame()

# Loop through each route in the data
for route in data['routes']:
    # Loop through each step in the route
    for step in route['steps']:
        # Normalize the step data (excluding 'fees' and 'gasCosts')
        step_data = {k: v for k, v in step.items() if k not in ['feeCosts', 'gasCosts']}
        step_df = pd.json_normalize(step_data)

        # Normalize and concatenate fees data if present
        if step['estimate']['feeCosts']:
            fees_df = pd.json_normalize(step['estimate']['feeCosts'])
            fees_df['type'] = 'feeCosts'  # Add a type column to distinguish fees
            step_df = pd.concat([step_df] * len(fees_df), ignore_index=True)
            step_df.update(fees_df)
        
        # Normalize and concatenate gas costs data if present
        if step['estimate']['gasCosts']:
            gas_costs_df = pd.json_normalize(step['estimate']['gasCosts'])
            gas_costs_df['type'] = 'gas_cost'  # Add a type column to distinguish gas costs
            step_df = pd.concat([step_df, pd.concat([step_df.iloc[0:1]] * len(gas_costs_df), ignore_index=True)], ignore_index=True)
            step_df.update(gas_costs_df, overwrite=False)
        
        # Extract the route metadata (excluding 'steps')
        metadata = {k: v for k, v in route.items() if k != 'steps'}
        metadata_df = pd.json_normalize(metadata)
        
        # Add a prefix to each metadata column name to prevent conflicts
        metadata_df.columns = ['route_' + col for col in metadata_df.columns]
        
        # Repeat the metadata for each row in the step_df
        repeated_metadata_df = pd.concat([metadata_df] * len(step_df), ignore_index=True)
        
        # Concatenate the step DataFrame with the repeated metadata DataFrame
        enriched_df = pd.concat([step_df.reset_index(drop=True), repeated_metadata_df], axis=1)
        
        # Append the enriched DataFrame to the normalized data DataFrame
        normalized_data_df = pd.concat([normalized_data_df, enriched_df], ignore_index=True)

# Display the DataFrame
print(normalized_data_df.head(15))

# Optionally, save the DataFrame to a CSV file
normalized_data_df.to_csv('enriched_normalized_routes_with_separate_fees_and_gas.csv', index=False)
```

```python
import pandas as pd
import json

# Load the JSON data from the file
file_path = 'routes/ETH-ARB-USDT.json'
with open(file_path, 'r') as file:
    data = json.load(file)

# Initialize an empty DataFrame for the normalized data
normalized_data_df = pd.DataFrame()

# Loop through each route in the data
for route in data['routes']:
    # Extract the route metadata (excluding 'steps')
    metadata = {k: v for k, v in route.items() if k != 'steps'}
    metadata_df = pd.json_normalize(metadata)
    metadata_df.columns = ['route_' + col for col in metadata_df.columns]  # Add a prefix to each metadata column name

    # Loop through each step in the route
    for step in route['steps']:
        # Normalize the step data (excluding 'feeCosts' and 'gasCosts')
        step_data = {k: v for k, v in step.items() if k not in ['feeCosts', 'gasCosts']}
        step_df = pd.json_normalize(step_data)
        step_df.columns = ['step_' + col for col in step_df.columns]  # Add a prefix to each metadata column name

        # Combine step data with route metadata
        combined_df = pd.concat([step_df] * len(metadata_df), ignore_index=True)
        combined_df = pd.concat([combined_df, metadata_df], axis=1)

        # Normalize and append fee costs data if present
        if step['estimate']['feeCosts']:
            for fee in step['estimate']['feeCosts']:
                fee_df = pd.json_normalize(fee)
                fee_df['cost_type'] = 'feeCost'  # Add a cost type column to distinguish fee costs
                fee_combined_df = pd.concat([combined_df.iloc[[0]], fee_df], axis=1)
                normalized_data_df = pd.concat([normalized_data_df, fee_combined_df], ignore_index=True)

        # Normalize and append gas costs data if present
        if step['estimate']['gasCosts']:
            for gas in step['estimate']['gasCosts']:
                gas_df = pd.json_normalize(gas)
                gas_df['cost_type'] = 'gasCost'  # Add a cost type column to distinguish gas costs
                gas_combined_df = pd.concat([combined_df.iloc[[0]], gas_df], axis=1)
                normalized_data_df = pd.concat([normalized_data_df, gas_combined_df], ignore_index=True)

        # If there are no feeCosts or gasCosts, just append the step data
        if not(step['estimate']['feeCosts']) and not(step['estimate']['gasCosts']):
            normalized_data_df = pd.concat([normalized_data_df, combined_df], ignore_index=True)

# Display the DataFrame
print(normalized_data_df.head())

# Optionally, save the DataFrame to a CSV file
normalized_data_df.to_csv('enriched_normalized_routes_with_separate_fees_and_gas.csv', index=False)
```
